<h1> Reading and Writing to Files using Libraries -> Pandas, JSON, CVS</h1>

<h2>1)  Modify a Spreadsheet for input into a Pandas Dataframe</h2>

```python
import pandas as pd
df = pd.read_csv('SavingsTable.csv')
print(df.to_string()) 
```

<h2>2.  Transform the Pandas Dataframe into a Dictionary</h2>

```python
savings=df.to_dict()
print(savings)
```

<h2>3.  Format the Dictionary for Display</h2>

```python
strHead="{:>12}"
y=[]
for x in savings:
    print(strHead.format(x),end="")
    y.append(savings[x])
print()
for row in range(len(y[0])):
    for col in range(len(y)):
        print(strHead.format(y[col][row]),end="")
    print()
```

<h2>4. Transpose the list and Display</h2>

```python
lsst=[]
for x in savings:
    lsst.append(x)
yi=[]
yi.append(lsst)
for row in range(len(y[0])):
    lst=[]
    for col in range(len(y)):
        lst.append(y[col][row])
    yi.append(lst)
print(yi)
```

<h2>5. Write the results to another .csv file</h2>

```python
import csv
f = open('SavingsTable2.csv', 'w', encoding='UTF8', newline='')
writer = csv.writer(f)
for row in yi:
    writer.writerow(row)
f.close()
```
